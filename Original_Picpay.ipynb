```python
# importando bibliotecas
import pandas as pd
import numpy as np
import openpyxl
from google.colab import files
```

```python
# leitura de dados
df_projuris = pd.read_csv('PROJURIS - RELATÓRIO CONFRONTO SEMANAL - 03-02-2025.csv', sep=';')
df_cpj = pd.read_csv('CPJ.csv', sep=';')
```

```python
# verificação de estrutura dos dados do cpj
df_cpj.shape
```

```python
# Função de tratamento dos dados dos numeros de processos
def tratar_processo(processo):
    # Remover "-" e "."
    processo = processo.replace('.', '').replace('-', '').replace('/','').replace('*', '')
    # Remover zeros à esquerda
    processo = processo.lstrip('0')
    return processo
```

```python
# Criar a coluna de processos tratados
df_projuris['numero_processo_tratado'] = df_projuris['Numero atual do processo'].apply(tratar_processo)
df_cpj['numero_processo_tratado'] = df_cpj['NUMERO PROCESSO'].apply(tratar_processo)
```

```python
# Verificação de consta ou não consta no Projuris
df_projuris['Consta no CPJ?'] = np.where(
    df_projuris['numero_processo_tratado'].isin(df_cpj['numero_processo_tratado']), 'Consta', 'Não consta'
)

df_projuris.head()
```

```python
# Verificação de consta ou não consta no CPJ
df_cpj['Consta no ProJuris?'] = np.where(
    df_cpj['numero_processo_tratado'].isin(df_projuris['numero_processo_tratado']), 'Consta', 'Não consta'
)

df_cpj.head()
```

```python
# Exportar os DataFrames para um arquivo Excel
arquivo_excel = 'processos_tratados.xlsx'
with pd.ExcelWriter(arquivo_excel) as writer:
    df_projuris.to_excel(writer, sheet_name='ProJuris', index=False)
    df_cpj.to_excel(writer, sheet_name='CPJ', index=False)

files.download(arquivo_excel)
```
